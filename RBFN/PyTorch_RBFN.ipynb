```python
# RBFN modeli oluşturmak ve RBFN Kernel'ın tanımlanması
```

```python
def rbf_kernel(X,centers,beta):
  return torch.exp(-beta*torch.cdist(X,centers)**2)
```

```python
class RBFN(nn.Module):
  def __init__(self,input_dim,num_centers,output_dim):
    super(RBFN,self).__init__() # Neural Network initiliazer çağırma işlemi
    self.centers = nn.Parameter(torch.randn(num_centers,input_dim)) # rbfn merkezlerini rastgele başlatma
    self.beta  = nn.Parameter(torch.ones(1)*2.0) # beta parametresi rbfn nin genişliğini kontrol edecek
    self.linear = nn.Linear(num_centers,output_dim) # outputu tam bağlantılı katmana yönlendir.
  def forward(self,x):
    # rbf çekirdek fonksiyonunu hesaplama
    phi = rbf_kernel(x,self.centers,self.beta)
    return self.linear(phi)
```

```python
num_centers = 10
model = RBFN(4,10,3)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(),lr=0.01)

num_epochs = 100
for epoch in range(num_epochs):
  optimizer.zero_grad()
  outputs = model(X_train)
  loss = criterion(outputs,y_train)
  loss.backward()
  optimizer.step()

  if (epoch+1) % 10 == 0:
    print(f"Epoch : {epoch+1}/{num_epochs} , Loss : {loss.item():.4f}")
```

```python
with torch.no_grad():
  y_pred = model(X_test) # test verisi ile tahmin
  accuracy = (torch.argmax(y_pred,axis=1)==y_test).float().mean().item()
  print(f"Accuracy : {accuracy}")
```
